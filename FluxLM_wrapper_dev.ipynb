## This code automates FLUX-LM, written in VBA, to solve for numerous temperature-depth profiles at different timesteps. The model estimates steady-state vertical groundwater flux in saturated conditions. ## 

The below cell opens the respective excel files and connects them with Python variables: One file contains the temperature-depth-time data that is copied, the other is the solver where the data is pasted to and Macros are solved.

```python
import win32com.client
import os
import openpyxl as xl
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

## VFLUX ##
user_input1 = input("Enter the path of your FLUX-LM file: ") #Path format example:
                                                          # 'C:\\Users\\cd34772\\Documents\\FLUX_LM\\Flux-LM'
EX=win32com.client.DispatchEx("Excel.Application")
os.getcwd()
EX.Visible=True
EX1=EX.Workbooks.Open(user_input1)
ws2 = EX1.worksheets(5)

#############

## WELL 1 DATA ##
user_input2 = input("Enter the path of your data file: ") #Path format example: 
                                                             #'C:\\Users\\cd34772\\Documents\\FLUX_LM\\Bredehoeft_table_3.xlsx'
BRED=win32com.client.DispatchEx("Excel.Application")
os.getcwd()
BRED.Visible=True
BRED1=BRED.Workbooks.Open(user_input2)
w1 = BRED1.worksheets[0]

##################

## WELL 2 DATA ##
os.getcwd()
w2 = BRED1.worksheets[1]
#################

## WELL 3 DATA ##
w3=BRED1.worksheets[2]
#################

# WELL 4 DATA ##
w4=BRED1.worksheets[3]
#################
```

Below 4 cells execute the copy/paste/run and stores the flux values in arrays.

```python
## WELL 1 RUN MACROS ##

user_input3 = input("Number of sensors for well 1: ")
user_input4 = input("Number of time steps for well 1 : ")
c1=np.zeros((1, user_input3), dtype=float)
d1=np.zeros((1, user_input3), dtype=float)
q1=np.zeros((1,user_input4), dtype=float)
E1=np.zeros((1,user_input4), dtype=float)

thickness1=w1.Cells(1,user_input3+1).Value
ws2.Cells(10,7).Value=thickness1

for m in range(0,user_input3):
    d1[0,m] = w1.Cells(1,m+2).Value
    ws2.Cells(m+10,1).Value=d1[0,m]
for j in range(2,user_input4+2):
    for i in range(0,user_input3):
        c1[0,i] = w1.Cells(j,i+2).Value
    for k in range(10,user_input3+10):        
        ws2.Cells(k,2).Value=c1[0,k-10]
    EX.Run("'Flux-LM.xlsm'!FluxOptimizer")
    q1[0,j-2]=ws2.Cells(3,4).Value
    E1[0,j-2]=ws2.Cells(4,4).Value
    
########################
```

```python
## WELL 2 RUN MACROS ##

user_input5 = input("Number of sensors for well 2: ")
user_input6 = input("Number of time steps for well 2 : ")
c2=np.zeros((1, user_input5), dtype=float)
d2=np.zeros((1, user_input5), dtype=float)
q2=np.zeros((1,user_input6), dtype=float)
E2=np.zeros((1,user_input6), dtype=float)

thickness2=w2.Cells(1,user_input5+1).Value
ws2.Cells(10,7).Value=thickness2

ws2.Cells(10,7).Value=thickness2
for m in range(0,user_input5):
    d2[0,m] = w2.Cells(1,m+2).Value
    ws2.Cells(m+10,1).Value=d2[0,m]
for j in range(2,user_input6+2):
    for i in range(0,user_input5):
        c2[0,i] = w2.Cells(j,i+2).Value
    for k in range(10,user_input5+10):
        ws2.Cells(k,2).Value=c2[0,k-10]
    EX.Run("'Flux-LM.xlsm'!FluxOptimizer")
    q2[0,j-2]=ws2.Cells(3,4).Value
    E2[0,j-2]=ws2.Cells(4,4).Value
    
########################

```

```python
## WELL 3 RUN MACROS ##

user_input7 = input("Number of sensors for well 3: ")
user_input8 = input("Number of time steps for well 3 : ")
c3=np.zeros((1, user_input7), dtype=float)
d3=np.zeros((1, user_input7), dtype=float)
q3=np.zeros((1,user_input8), dtype=float)
E3=np.zeros((1,user_input8), dtype=float)

thickness3=w3.Cells(1,user_input7+1).Value
ws2.Cells(10,7).Value=thickness3 # assign total thickness (d)

for m in range(0,user_input7):
    d3[0,m] = w3.Cells(1,m+2).Value
    ws2.Cells(m+10,1).Value=d3[0,m]
for j in range(2,user_input8+2):
    for i in range(0,user_input7):
        c3[0,i] = w3.Cells(j,i+2).Value
    for k in range(10,user_input7+10):
        ws2.Cells(k,2).Value=c3[0,k-10]
    EX.Run("'Flux-LM.xlsm'!FluxOptimizer")
    q3[0,j-2]=ws2.Cells(3,4).Value
    E3[0,j-2]=ws2.Cells(4,4).Value
    
########################

```

```python
## WELL 4 RUN MACROS ##

user_input9 = input("Number of sensors for well 4: ")
user_input10 = input("Number of time steps for well 4 : ")
c4=np.zeros((1, user_input9), dtype=float)
d4=np.zeros((1, user_input9), dtype=float)
q4=np.zeros((1,user_input10), dtype=float)
E4=np.zeros((1,user_input10), dtype=float)

thickness4=w4.Cells(1,user_input9+1).Value
ws2.Cells(10,7).Value=thickness4
for m in range(0,user_input9):
    d4[0,m] = w4.Cells(1,m+2).Value
    ws2.Cells(m+10,1).Value=d4[0,m]
for j in range(2,user_input10+2):
    for i in range(0,user_input9):
        c4[0,i] = w4.Cells(j,i+2).Value
    #print(c)
    for k in range(10,user_input9+10):
        ws2.Cells(k,2).Value=c4[0,k-10]
    EX.Run("'Flux-LM.xlsm'!FluxOptimizer")
    q4[0,j-2]=ws2.Cells(3,4).Value
    E4[0,j-2]=ws2.Cells(4,4).Value

#######################
```
